# Data visualization with matplotlib

*Matplotlib* is a python package for data visualization. It can generate several chart formats like line charts and scatterplots.

```python
# First, make sure that you have a working installation of matplotlib.
import matplotlib.pyplot as plt
```

```python
# Run this cell to make the Jupyter notebook show the generated charts in the cell output
%matplotlib inline
```

## Line charts

Line charts are used to plot datapoints connected with a straight line. They're useful to see trends in the relationship between two ordered sequences of values(time-series data, usually).

```python
x = range(10)
y = [x_val**2 for x_val in x]

plt.plot(x, y)
plt.show()
```

*Be careful about differences in axis ranges!*

## Scatter plots

Scatter plots shows data points corresponding to pairs of values. They're useful when you need to see the relationship between two **unordered** sets of values.

```python
# Let's start by retrieving the Boston housing dataset
import pandas as pd

boston_housing = pd.read_csv('datasets/boston.csv')
```

```python
boston_housing.rm.sort_values()
```

```python
plt.scatter(boston_housing.rm, boston_housing.medv)
plt.show()
```

## Histograms

Histograms are used to visualize the distribution of values in a given data collection. They allow us to see the overall "shape" of our data.

```python
plt.hist(boston_housing.medv)
plt.show()
```

## Chart customization

Matplotlib charts can be customized with labels, axis dimensions, colors, shapes, etc.

```python
# A list of supported markers can be found in http://matplotlib.org/api/markers_api.html
plt.scatter(boston_housing.rm, boston_housing.medv, color='green', marker='o', facecolors='none')
plt.title('Number of rooms x House prices')
plt.xlabel('Number of rooms')
plt.ylabel('Price in 1000$')
plt.show()
```

## Superimposition of charts

It's possible to plot more than one chart in the same figure using matplotlib. To achieve this, make the chart generation calls(plt.plot, plo.scatter...) before calling plt.show().

```python
plt.title('Number of rooms x House prices')
plt.xlabel('Number of rooms')
plt.ylabel('Price in 1000$')
plt.scatter(boston_housing.rm, boston_housing.medv, color='blue', marker='o', facecolors='none')
plt.show()
```

```python
rm_line = [(12*x_val-55) for x_val in boston_housing.rm]
plt.plot(boston_housing.rm, rm_line, color='red')
```

```python
plt.title('Number of rooms x House prices')
plt.xlabel('Number of rooms')
plt.ylabel('Price in 1000$')
plt.scatter(boston_housing.rm, boston_housing.medv, color='blue', marker='o', facecolors='none')
plt.plot(boston_housing.rm, rm_line, color='red')
plt.show()
```
